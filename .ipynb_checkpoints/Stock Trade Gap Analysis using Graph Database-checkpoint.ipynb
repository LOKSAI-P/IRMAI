```python
import gradio as gr
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from neo4j import GraphDatabase
from fpdf import FPDF
import tempfile

# Connect to Neo4j database
NEO4J_URL = "bolt://localhost:7687"  
NEO4J_USERNAME = "neo4j"
NEO4J_PASSWORD = "password"

neo4j_driver = GraphDatabase.driver(NEO4J_URL, auth=(NEO4J_USERNAME, NEO4J_PASSWORD))
print("Connected to Neo4j successfully!")

# Generate sample stock trade data with gaps
np.random.seed(42)
trade_data = {
    "trade_id": range(1, 21),
    "ticker": ["AAPL"] * 10 + ["MSFT"] * 10,
    "volume": [1000, 1200, 0, 1100, 1300, 1400, 1250, 0, 1150, 1350,
               2000, 0, 2100, 2500, 2600, 2700, 2550, 2300, 0, 2450],
    "price": [150, 152, 180, 151, 153, 154, 149, 0, 150, 155,
              300, 0, 305, 310, 315, 318, 312, 290, 0, 295],
    "timestamp": pd.date_range("2024-02-01", periods=20, freq="h")
}

trade_df = pd.DataFrame(trade_data)

# Insert trades into Neo4j
def insert_trade_node(tx, trade):
    query = """
    CREATE (t:Trade {trade_id: $trade_id, ticker: $ticker,
                     volume: $volume, price: $price, timestamp: $timestamp})
    """
    tx.run(query, **trade)

with neo4j_driver.session() as session:
    for _, row in trade_df.iterrows():
        session.write_transaction(insert_trade_node, row.to_dict())

print("Inserted stock trades into Neo4j!")

# Create relationships between trades
def create_trade_relationships(tx):
    query = """
    MATCH (t1:Trade), (t2:Trade) 
    WHERE t1.timestamp < t2.timestamp
    CREATE (t1)-[:NEXT_TRADE]->(t2)
    """
    tx.run(query)

with neo4j_driver.session() as session:
    session.write_transaction(create_trade_relationships)

print("Created relationships between consecutive trades!")

# Identify gaps in trade data
trade_df["missing_volume"] = trade_df["volume"] == 0
trade_df["missing_price"] = trade_df["price"] == 0
trade_df["price_change"] = trade_df["price"].diff().fillna(0).abs()
trade_df["sudden_jump"] = trade_df["price_change"] > 20
trade_df["is_gap"] = trade_df["missing_volume"] | trade_df["missing_price"] | trade_df["sudden_jump"]

# Mark gaps in Neo4j
def mark_trade_gaps(tx, trade_id, is_gap):
    query = """
    MATCH (t:Trade {trade_id: $trade_id})
    SET t.is_gap = $is_gap
    """
    tx.run(query, trade_id=int(trade_id), is_gap=bool(is_gap))

with neo4j_driver.session() as session:
    for _, row in trade_df.iterrows():
        session.write_transaction(mark_trade_gaps, row["trade_id"], row["is_gap"])

print("Marked gaps in Neo4j!")

# Function to plot graphs
def plot_trade_graphs():
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    # Stock Price Trends with Gaps Highlighted
    colors = trade_df["is_gap"].map({True: "red", False: "green"})
    axes[0].scatter(trade_df["timestamp"], trade_df["price"], c=colors)
    axes[0].plot(trade_df["timestamp"], trade_df["price"], linestyle="-", marker="o", alpha=0.6)
    axes[0].set_xlabel("Timestamp")
    axes[0].set_ylabel("Price")
    axes[0].set_title("Stock Price Trends with Gaps")
    axes[0].grid(True)

    # Trade Network Graph
    trade_network = nx.DiGraph()
    
    for _, row in trade_df.iterrows():
        node_color = "red" if row["is_gap"] else "green"
        trade_network.add_node(row["trade_id"], color=node_color)
        
    for i in range(len(trade_df) - 1):
        trade_network.add_edge(trade_df.loc[i, "trade_id"], trade_df.loc[i + 1, "trade_id"])
        
    pos = nx.spring_layout(trade_network)
    node_colors = [trade_network.nodes[n]["color"] for n in trade_network.nodes]
    
    nx.draw(trade_network,
            pos,
            with_labels=True,
            node_color=node_colors,
            edge_color="gray",
            node_size=800,
            font_size=10,
            font_weight="bold",
            arrowsize=12,
            ax=axes[1])
            
    axes[1].set_title("Stock Trade Network Graph")

    plt.tight_layout()
    
    return fig

# Function to generate PDF report with summary and graphs
def generate_pdf_report():
    pdf_report = FPDF()
    pdf_report.set_auto_page_break(auto=True, margin=15)

    # Title page
    pdf_report.add_page()
    pdf_report.set_font("Arial", "B", 14)
    pdf_report.cell(200, 10, "Stock Trade Analysis Report", ln=True, align="C")
    
    # Summary Section
    pdf_report.set_font("Arial", "", 11)
    pdf_report.cell(0, 8, f"Total Trades: {len(trade_df)}", ln=True)
    pdf_report.cell(0, 8, f"Total Gaps Detected: {trade_df['is_gap'].sum()}", ln=True)
    pdf_report.cell(0, 8, f"Missing Volume Entries: {trade_df['missing_volume'].sum()}", ln=True)
    pdf_report.cell(0, 8, f"Missing Price Entries: {trade_df['missing_price'].sum()}", ln=True)
    
    # Create a new page for graphs
    pdf_report.add_page()

    # Save and add Stock Price Trends graph
    price_trends_fig = plt.figure(figsize=(6, 4))
    
    colors = trade_df["is_gap"].map({True: "red", False: "green"})
    
    plt.scatter(trade_df["timestamp"], trade_df["price"], c=colors)
    plt.plot(trade_df["timestamp"], trade_df["price"], linestyle="-", marker="o", alpha=0.6)
    
    plt.xlabel("Timestamp")
    plt.ylabel("Price")
    plt.title("Stock Price Trends with Gaps")
    
    price_trends_path = tempfile.NamedTemporaryFile(delete=False).name + ".png"
    
    plt.savefig(price_trends_path)
    
    plt.close(price_trends_fig)

   # Save and add Trade Network Graph 
   network_fig = plt.figure(figsize=(6, 4))

   network_graph = nx.DiGraph()
   for _, row in trade_df.iterrows():
       node_color = "red" if row["is_gap"] else "green"
       network_graph.add_node(row["trade_id"], color=node_color)

   for i in range(len(trade_df) - 1):
       network_graph.add_edge(trade_df.loc[i]["trade_id"], trade_df.loc[i + 1]["trade_id"])

   pos = nx.spring_layout(network_graph)

   node_colors = [network_graph.nodes[n]["color"] for n in network_graph.nodes]

   nx.draw(network_graph,
           pos,
           with_labels=True,
           node_color=node_colors,
           edge_color="gray",
           node_size=500,
           font_size=8,
           font_weight="bold",
           arrowsize=10)

   network_path = tempfile.NamedTemporaryFile(delete=False).name + ".png"
   plt.savefig(network_path)
   plt.close(network_fig)

   # Add both images on the same page of the PDF report 
   pdf_report.image(price_trends_path , x=10 , y=30 , w=90)  
   pdf_report.image(network_path , x=110 , y=30 , w=90)  
   
   # Add Data Table Page 
   pdf_report.add_page()
   pdf_report.set_font("Arial", "B",9 )
   
   pdf_report.cell(0 ,8 , "Trade Data Table:" , ln=True )
   
   pdf_report.set_font("Arial", "",9 )  

   column_width =25  
   headers = trade_df.columns.tolist()

   for header in headers:
       pdf_report.cell(column_width ,8 , header , border=1 , align="C")

   pdf_report.ln()

   for _, row in trade_df.iterrows():
       for value in row:
           pdf_report.cell(column_width ,8 , str(value) , border=1 , align="C")
       pdf_report.ln()

   # Save PDF report 
   pdf_path = tempfile.NamedTemporaryFile(delete=False).name + ".pdf"
   pdf_report.output(pdf_path)

   return pdf_path

# Gradio Interface Setup 
graph_component = gr.Plot(plot_trade_graphs)
table_component = gr.DataFrame(trade_df)
pdf_download_component = gr.File(generate_pdf_report , label="Download Report")

demo_interface = gr.Interface(
     fn=None ,
     inputs=[] ,
     outputs=[graph_component , table_component , pdf_download_component],
     live=False ,
     title="Stock Trade Gap Analysis",
     description="Analyze stock trades for gaps with graphs and downloadable reports."
)

demo_interface.launch()
```
